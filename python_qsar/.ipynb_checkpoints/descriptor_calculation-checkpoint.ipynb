# The feature selection was performed with Mordred. A specific version of NumPy was necessary. Consequently, a separate jupyter notebook in a mordred environment (mordred_env) was created.

### Plot Feature Scores for Overview

```python
np.random.seed(42)
dataset = pd.read_csv('./datasets/x_inhousechembl_desc.csv')
target = pd.read_csv('./datasets/y_inhousechembl.csv')
x_values = dataset.values
y_values = target.values.ravel()

#Get column names of selected features
feature_names = list(dataset.columns.values)

x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.2, random_state=42)
 
# feature selection
x_train_fs, x_test_fs, fs = select_features(x_train, y_train, x_test, num_features=25)
# what are scores for the features
for i in range(len(fs.scores_)):
	print('%s: %f' % (feature_names[i], fs.scores_[i]))
# plot the scores
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.xlabel('Feature ID',fontproperties = 'Times New Roman', size = 16)
plt.ylabel('Mutual Information',fontproperties = 'Times New Roman', size = 16)

#save figure
plt.savefig(".\\figures\\inhousechembl_feature_scores.png", bbox_inches='tight', pad_inches= 0)

plt.show()
count=0
for score in fs.scores_:
    if score >= 0.5:
        count= count+1
print(f'Scores >= 0.5: {count}')


#write feature scores to csv
feature_scores = np.stack((feature_names, fs.scores_), axis=1)
feature_scores.tofile('.\\datasets\\inhousechembl_feature_scores.csv', sep = ',')
```

### Normalize descriptor set and select best 25 features

```python
np.random.seed(42)
dataset = pd.read_csv('./datasets/x_inhousechembl_desc.csv')
target = pd.read_csv('./datasets/y_inhousechembl.csv')
x_values = dataset.values
y_values = target.values.ravel()

x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.2, random_state=42)

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

x_train_inhousechembl_desc25, x_test_inhousechembl_desc25, fs = select_features(x_train, y_train, x_test, num_features=25)

#Get column names of selected features
column_names = list(dataset.columns.values)
column_names = fs.get_feature_names_out(input_features=column_names)

#Save feature list as json
with open(r'settings\best_descriptors.json', 'w') as file:
    json.dump(list(column_names), file)
    
#Save datasets with column names
pd.DataFrame(x_train_inhousechembl_desc25).set_axis(column_names, axis=1).to_csv(r'.\datasets\x_train_inhousechembl_desc25.csv', index = False)
pd.DataFrame(x_test_inhousechembl_desc25).set_axis(column_names, axis=1).to_csv(r'.\datasets\x_test_inhousechembl_desc25.csv', index = False)
```

## Train test split of 20 selected descriptors for In-house-ChEMBL without Zeros
Not all 25 best descriptors could be calculated for all compounds of all databases. The 20 best and consistently calculable descriptors were selected.

```python
descriptor_list = ["nN", "ATS8m", "ATS3v", "ATS6i",  "BCUTse-1l", "BCUTpe-1l", "BCUTare-1l", "nBondsD", "Xc-3d", 
                   "NdO", "SssNH", "SdO", "ETA_beta_s", "TIC1", "SMR_VSA3", "SMR_VSA6", "SlogP_VSA1", "MID_N", "piPC4", "MWC08"]
     
# Create empty Calculator instance
calc = Calculator()

# Register descriptor instance
calc.register(AtomCount.AtomCount(type='N'))
calc.register(Autocorrelation.ATS(order=8, prop='m'))
calc.register(Autocorrelation.ATS(order=3, prop='v'))
calc.register(Autocorrelation.ATS(order=6, prop='i'))
calc.register(BCUT.BCUT(prop='se', nth=-1))
calc.register(BCUT.BCUT(prop='pe', nth=-1))
calc.register(BCUT.BCUT(prop='are', nth=-1))
calc.register(BondCount.BondCount(type='double', kekulize=False))
calc.register(Chi.Chi(type='cluster', order=3, prop='d', averaged=False))
calc.register(EState.AtomTypeEState(type='count', estate='dO'))
calc.register(EState.AtomTypeEState(type='sum', estate='ssNH'))
calc.register(EState.AtomTypeEState(type='sum', estate='dO'))
calc.register(ExtendedTopochemicalAtom.EtaVEMCount(type='s', averaged=False))
calc.register(InformationContent.TotalIC(order=1))
calc.register(MoeType.SMR_VSA(k=3))
calc.register(MoeType.SMR_VSA(k=6))
calc.register(MoeType.SlogP_VSA(k=1))
calc.register(MolecularId.MolecularId(type='N', averaged=False, _eps=1e-10))
calc.register(PathCount.PathCount(order=4, pi=True, total=False, log=True))
calc.register(WalkCount.WalkCount(order=8, total=False, self_returning=False))

# Calculate descriptors for In-house+ChEMBL without Zeros and apply MinMaxScaler.

np.random.seed(42)

# Get pChEMBL values
target = pd.read_csv('./datasets/y_inhousechembl.csv')

# Calculate 20 selected descriptors
file_name = '.\\datasets\\input\\inhouse.csv'
df_smiles = pd.read_csv(file_name, delimiter=',')    
smiles = df_smiles['smiles']
smiles_list = np.array(smiles).tolist()
mols = [Chem.MolFromSmiles(smile) for smile in smiles_list]
df_desc = calc.pandas(mols)
df_desc = drop_nonnumeric_columns(df_desc)

#Save datasets with column names
df_desc.set_axis(descriptor_list, axis=1).to_csv(r'.\datasets\x_inhousechembl_desc20.csv', index = False)
```

### Calculate 20 Selected Mordred Descriptors for all Databases

```python
database_list = ['bms300k', 'chembl34', 'clickcyctetpep', 'coconut', 
                 'cyctetpep', 'divset', 'enaminescreening', 
                 'enaminePPI', 'iPPI', 'npatlas', 'reinventPPI', 'chemdivkras']
descriptor_list = ["nN", "ATS8m", "ATS3v", "ATS6i",  "BCUTse-1l", "BCUTpe-1l", "BCUTare-1l", "nBondsD", "Xc-3d", 
                   "NdO", "SssNH", "SdO", "ETA_beta_s", "TIC1", "SMR_VSA3", "SMR_VSA6", "SlogP_VSA1", "MID_N", "piPC4", "MWC08"]
dropped_desc_list = ["ATS0s", "ATS2s", "ATS5s", "ATS6s","BCUTs-1l",]   
dropped_mols_dict = {}

# Create empty Calculator instance
calc = Calculator()

# Register descriptor instance
calc.register(AtomCount.AtomCount(type='N'))
#calc.register(Autocorrelation.ATS(order=0, prop='s'))
#calc.register(Autocorrelation.ATS(order=2, prop='s'))
#calc.register(Autocorrelation.ATS(order=5, prop='s'))
#calc.register(Autocorrelation.ATS(order=6, prop='s'))
calc.register(Autocorrelation.ATS(order=8, prop='m'))
calc.register(Autocorrelation.ATS(order=3, prop='v'))
calc.register(Autocorrelation.ATS(order=6, prop='i'))
#calc.register(BCUT.BCUT(prop='s', nth=-1))
calc.register(BCUT.BCUT(prop='se', nth=-1))
calc.register(BCUT.BCUT(prop='pe', nth=-1))
calc.register(BCUT.BCUT(prop='are', nth=-1))
calc.register(BondCount.BondCount(type='double', kekulize=False))
calc.register(Chi.Chi(type='cluster', order=3, prop='d', averaged=False))
calc.register(EState.AtomTypeEState(type='count', estate='dO'))
calc.register(EState.AtomTypeEState(type='sum', estate='ssNH'))
calc.register(EState.AtomTypeEState(type='sum', estate='dO'))
calc.register(ExtendedTopochemicalAtom.EtaVEMCount(type='s', averaged=False))
calc.register(InformationContent.TotalIC(order=1))
calc.register(MoeType.SMR_VSA(k=3))
calc.register(MoeType.SMR_VSA(k=6))
calc.register(MoeType.SlogP_VSA(k=1))
calc.register(MolecularId.MolecularId(type='N', averaged=False, _eps=1e-10))
calc.register(PathCount.PathCount(order=4, pi=True, total=False, log=True))
calc.register(WalkCount.WalkCount(order=8, total=False, self_returning=False))

for database in database_list:
    dropped_mols_list = []
    file_name = '.\\datasets\\input\\'+ database +'.csv'
    df_smiles = pd.read_csv(file_name, delimiter=',')    
    smiles = df_smiles['smiles']
    smiles_list = np.array(smiles).tolist()
    mols = [Chem.MolFromSmiles(smile) for smile in smiles_list]
    for idx, mol in enumerate(mols):
        if mol == None:
            mols.pop(idx)
            dropped_mols_list.append(smiles_list[idx])
    dropped_mols_dict[database] = dropped_mols_list
    df_desc = calc.pandas(mols)
    df_desc = drop_nonnumeric_columns(df_desc)   
    df_desc.set_axis(descriptor_list, axis=1).to_csv('.\\datasets\\x_'+ database +'_desc20.csv', index = False)

with open(".\\settings\\dropped_mols.json", 'w') as json_file:
    json.dump(dropped_mols_dict, json_file, indent=4)
```

### Not all molecules could be created from smiles in the ChEMBL34 and Coconut libraries. Molecules which could not be calculated were removed here.

```python
# Attempt to create molecules from each smiles in ChEMBL34. Empty results were collected in a list.
dropped_mols_list = []
file_name = '.\\datasets\\input\\chembl34.csv'
df_smiles = pd.read_csv(file_name, delimiter=',')    
smiles = df_smiles['smiles']
smiles_list = np.array(smiles).tolist()
mols = [Chem.MolFromSmiles(smile) for smile in smiles_list]
for idx, mol in enumerate(mols):
    if mol == None:
        mols.pop(idx)
        dropped_mols_list.append(smiles_list[idx])
```

```python
# Remove all smiles of dropped molecules from ChEMBL34
for mol in dropped_mols_list:
    mask = df_smiles['smiles'] == mol
    df_smiles = df_smiles[~mask]
df_smiles.shape
```

```python
# Save the reduced ChEMBL34 library
df_smiles.to_csv('.\\datasets\\input\\chembl34_dropped.csv', index = False)
```

```python
"""Repeat the reduction process with the coconut library."""

dropped_mols_list = []
file_name = '.\\datasets\\input\\coconut.csv'
df_smiles = pd.read_csv(file_name, delimiter=',')    
smiles = df_smiles['smiles']
smiles_list = np.array(smiles).tolist()
mols = [Chem.MolFromSmiles(smile) for smile in smiles_list]
for idx, mol in enumerate(mols):
    if mol == None:
        mols.pop(idx)
        dropped_mols_list.append(smiles_list[idx])
for mol in dropped_mols_list:
    mask = df_smiles['smiles'] == mol
    df_smiles = df_smiles[~mask]
df_smiles.to_csv('.\\datasets\\input\\coconut_dropped.csv', index = False)
```
